```python
before = len(df)
df.dropna(how='all', inplace=True)
print(f"Dropped {before - len(df)} fully-empty rows. Remaining: {len(df):,}")
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.title()                          # keeps Power BI column names readable
    .str.replace(r'\s+', ' ', regex=True) # collapse double spaces
)
print("Columns:", list(df.columns))
```

```python
# Ports and Attack ID were loaded as float64 because of nulls — convert to Int64 (nullable int)
for col in ['Attack Id', 'Source Port', 'Destination Port']:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')

# Payload Size and Confidence Score stay as float
for col in ['Payload Size (Bytes)', 'Confidence Score']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.dtypes
```

```python
# ── Categorical columns → fill with 'Unknown'
cat_cols = [
    'Source Ip', 'Destination Ip',
    'Source Country', 'Destination Country',
    'Protocol', 'Attack Type', 'Detection Label',
    'Ml Model', 'Affected System', 'Port Type'
]
for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown').str.strip()

# ── Numeric columns → fill with median
num_cols = ['Source Port', 'Destination Port', 'Payload Size (Bytes)', 'Confidence Score']
for col in num_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

print("Nulls remaining:")
df.isnull().sum()
```

```python
# Timestamp has ~90k nulls — parse what we can, leave the rest as NaT
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')

# Extract useful parts for analysis
df['Attack Date']  = df['Timestamp'].dt.date
df['Attack Year']  = df['Timestamp'].dt.year
df['Attack Month'] = df['Timestamp'].dt.month_name()
df['Attack Hour']  = df['Timestamp'].dt.hour
df['Day Of Week']  = df['Timestamp'].dt.day_name()

print(f"Valid timestamps: {df['Timestamp'].notna().sum():,}")
print(f"Missing timestamps (NaT): {df['Timestamp'].isna().sum():,}")
```

```python
# Binary detection flag (easier for charts)
df['Is Detected'] = df['Detection Label'].str.strip().str.lower().map(
    {'detected': 1, 'not detected': 0}
).fillna(-1).astype(int)

# Payload size bucket
df['Payload Bucket'] = pd.cut(
    df['Payload Size (Bytes)'],
    bins=[0, 1_000, 10_000, 50_000, 100_000, float('inf')],
    labels=['< 1 KB', '1–10 KB', '10–50 KB', '50–100 KB', '> 100 KB']
)

# Confidence tier
df['Confidence Tier'] = pd.cut(
    df['Confidence Score'],
    bins=[0, 0.4, 0.7, 1.01],
    labels=['Low', 'Medium', 'High']
)

df[['Detection Label', 'Is Detected', 'Payload Bucket', 'Confidence Tier']].head(10)
```

```python
before = len(df)
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print(f"Duplicates removed: {before - len(df)}")
print(f"Final shape: {df.shape[0]:,} rows × {df.shape[1]} cols")
```
